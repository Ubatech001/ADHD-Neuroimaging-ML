```python
# NLP Analysis on Brown Corpus
# This notebook performs text analysis including TF, TF-IDF, POS tagging, and trigram analysis

# Import required libraries
import nltk
from nltk.corpus import brown
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer
from nltk import pos_tag
from nltk.util import trigrams
from collections import Counter, defaultdict
import string
import math
```

```python
# Download required NLTK data
nltk.download('brown')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab') # Added to ensure punkt_tab is downloaded
```

```python
# ============================================================================
# Task 1: Load the Brown Corpus using paras()
# ============================================================================
print("=" * 60)
print("TASK 1: Loading Brown Corpus paragraphs")
print("=" * 60)

# Load all paragraphs from Brown corpus
paragraphs = list(brown.paras())
print(f"Loaded {len(paragraphs)} paragraphs")

# Flatten paragraphs to get a list of sentences for easier processing
# Each paragraph is a list of sentences, each sentence is a list of words
documents = []
for paragraph in paragraphs:
    # Combine sentences in each paragraph into a single document
    doc = []
    for sentence in paragraph:
        doc.extend(sentence)
    documents.append(' '.join(doc))

print(f"Created {len(documents)} documents (one per paragraph)")
```

```python
# ============================================================================
# Task 2: Remove punctuation and stopwords, apply Lancaster stemmer
# ============================================================================
print("\n" + "=" * 60)
print("TASK 2: Text preprocessing (remove punctuation, stopwords, apply stemming)")
print("=" * 60)

nltk.download('punkt_tab') # Added to resolve LookupError for 'punkt_tab'

# Initialize stemmer and stopwords
stemmer = LancasterStemmer()
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Preprocess each document
preprocessed_docs = []
all_stems = []  # For TF calculation

for doc in documents:
    # Tokenize
    tokens = word_tokenize(doc.lower())

    # Remove punctuation and stopwords, apply stemming
    processed_tokens = []
    for token in tokens:
        if token not in punctuation and token not in stop_words:
            # Stem the token
            stemmed = stemmer.stem(token)
            processed_tokens.append(stemmed)
            all_stems.append(stemmed)

    preprocessed_docs.append(processed_tokens)

print(f"Preprocessed {len(preprocessed_docs)} documents")
print(f"Total tokens after preprocessing: {len(all_stems)}")
```

```python
# ============================================================================
# Task 3: Print top 10 words in terms of TF with their TF values
# ============================================================================
print("\n" + "=" * 60)
print("TASK 3: Top 10 words by Term Frequency (TF)")
print("=" * 60)

# Calculate term frequencies across all documents
tf_counter = Counter(all_stems)
total_tokens = len(all_stems)

# Get top 10 terms with their TF values
top_10_tf = tf_counter.most_common(10)

print(f"\n{'Word':<15} {'Frequency':<12} {'TF':<10}")
print("-" * 40)
for word, freq in top_10_tf:
    tf_value = freq / total_tokens
    print(f"{word:<15} {freq:<12} {tf_value:.6f}")
```

```python
# ============================================================================
# Task 4: Print top 10 words in terms of TF-IDF
# ============================================================================
print("\n" + "=" * 60)
print("TASK 4: Top 10 words by TF-IDF (using paragraphs as documents)")
print("=" * 60)

# Calculate document frequencies (number of documents containing each term)
doc_freq = defaultdict(int)
for doc in preprocessed_docs:
    unique_terms = set(doc)
    for term in unique_terms:
        doc_freq[term] += 1

# Calculate TF-IDF for each term
N = len(preprocessed_docs)  # Total number of documents
tfidf_scores = defaultdict(float)

for term in tf_counter:
    # Calculate TF (term frequency)
    tf = tf_counter[term] / total_tokens

    # Calculate IDF (inverse document frequency)
    idf = math.log(N / (1 + doc_freq[term]))

    # Calculate TF-IDF
    tfidf = tf * idf
    tfidf_scores[term] = tfidf

# Get top 10 terms by TF-IDF
top_10_tfidf = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:10]

print(f"\n{'Word':<15} {'TF-IDF Score':<15}")
print("-" * 30)
for word, score in top_10_tfidf:
    print(f"{word:<15} {score:.6f}")
```

```python
# ============================================================================
# Task 5: POS tagging and trigrams of word-tag pairs
# ============================================================================
print("\n" + "=" * 60)
print("TASK 5: Top 10 most common trigrams of word-tag pairs")
print("=" * 60)

# Process original documents for POS tagging (use original text, not stemmed)
all_pos_tagged = []

for doc in documents:
    # Tokenize original document
    tokens = word_tokenize(doc)

    # Apply POS tagging
    tagged_tokens = pos_tag(tokens)

    # Create word-tag pairs (convert to lowercase for consistency)
    word_tag_pairs = [(word.lower(), tag) for word, tag in tagged_tokens]
    all_pos_tagged.extend(word_tag_pairs)

# Generate trigrams from word-tag pairs
pos_trigrams = list(trigrams(all_pos_tagged))

# Count trigram frequencies
trigram_counter = Counter(pos_trigrams)

# Get top 10 trigrams
top_10_trigrams = trigram_counter.most_common(10)

print(f"\n{'Trigram (word-tag pairs)':<50} {'Frequency':<10}")
print("-" * 65)
for trigram, freq in top_10_trigrams:
    # Format trigram for display
    formatted_trigram = " -> ".join([f"{word}/{tag}" for word, tag in trigram])
    print(f"{formatted_trigram:<50} {freq:<10}")

print("\n" + "=" * 60)
print("Analysis completed successfully!")
print("=" * 60)
```
